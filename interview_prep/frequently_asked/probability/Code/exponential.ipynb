```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
lambda_param = 0.5  # rate parameter
n_samples = 1000    # number of samples
x = np.linspace(0, 10, 100)  # for plotting PDF

# Generate random samples from exponential distribution
samples = np.random.exponential(scale=1/lambda_param, size=n_samples)

# Calculate PDF
pdf = lambda_param * np.exp(-lambda_param * x)

# Create plots
plt.figure(figsize=(12, 6))

# Plot 1: Histogram of samples with PDF overlay
plt.subplot(1, 2, 1)
plt.hist(samples, bins=30, density=True, alpha=0.7, label='Histogram of Samples')
plt.plot(x, pdf, 'r-', lw=2, label=f'PDF (λ={lambda_param})')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Exponential Distribution: Histogram and PDF')
plt.legend()
plt.grid(True)

# Plot 2: Different rate parameters
plt.subplot(1, 2, 2)
for lambda_val in [0.5, 1.0, 2.0]:
    pdf = lambda_val * np.exp(-lambda_val * x)
    plt.plot(x, pdf, lw=2, label=f'λ={lambda_val}')

plt.xlabel('x')
plt.ylabel('Density')
plt.title('Exponential PDFs with Different Rate Parameters')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate sample statistics
print(f"Sample Statistics:")
print(f"Mean (theoretical = {1/lambda_param:.3f}):")
print(f"Sample mean = {np.mean(samples):.3f}")
print(f"\nVariance (theoretical = {1/lambda_param**2:.3f}):")
print(f"Sample variance = {np.var(samples):.3f}")
```
