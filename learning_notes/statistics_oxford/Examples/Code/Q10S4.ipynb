```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.special import gamma as gamma_func

# Set up the plotting parameters
plt.style.use('seaborn-v0_8')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Define the range of theta values to plot
theta = np.linspace(0, 5, 1000)

# Prior parameters
alpha1, beta1 = 4, 23    # Case (i)
alpha2, beta2 = 36, 6    # Case (ii)

# Calculate mean and variance for both priors
mean1 = alpha1 / beta1
var1 = alpha1 / (beta1**2)
mean2 = alpha2 / beta2
var2 = alpha2 / (beta2**2)

# Calculate the prior densities
prior1 = (beta1**alpha1 / gamma_func(alpha1)) * theta**(alpha1-1) * np.exp(-beta1*theta)
prior2 = (beta2**alpha2 / gamma_func(alpha2)) * theta**(alpha2-1) * np.exp(-beta2*theta)

# Plot the first prior
ax1.plot(theta, prior1, 'b-', label=f'$={alpha1}, ={beta1}$', linewidth=2)
ax1.axvline(mean1, color='r', linestyle='--', label=f'Mean={mean1:.2f}')
ax1.axvline(2, color='g', linestyle='--', label='$=2$')
ax1.fill_between(theta, prior1, alpha=0.3)
ax1.set_title('Prior Distribution (Case i)',fontsize=14)
ax1.set_xlabel('$$',fontsize=14)
ax1.set_ylabel('Density',fontsize=14)
ax1.legend(fontsize=18)
ax1.grid(True)

# Plot the second prior
ax2.plot(theta, prior2, 'b-', label=f'$={alpha2}, ={beta2}$', linewidth=2)
ax2.axvline(mean2, color='r', linestyle='--', label=f'Mean={mean2:.2f}')
ax2.axvline(2, color='g', linestyle='--', label='$=2$')
ax2.fill_between(theta, prior2, alpha=0.3)
ax2.set_title('Prior Distribution (Case ii)',fontsize=14)
ax2.set_xlabel('$$',fontsize=14)
ax2.set_ylabel('Density',fontsize=14)
ax2.legend(fontsize=18)
ax2.grid(True)

# Calculate and print summary statistics
print(f"Case (i) - Mean: {mean1:.2f}, Variance: {var1:.2f}, Mode: {(alpha1-1)/beta1:.2f}")
print(f"Case (ii) - Mean: {mean2:.2f}, Variance: {var2:.2f}, Mode: {(alpha2-1)/beta2:.2f}")

plt.tight_layout()
plt.show()
```
